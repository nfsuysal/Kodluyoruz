```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans

def k_means(k,l):
    kmeans = KMeans(n_clusters=k, random_state=0).fit(l)
    kmeans.fit(l) 
    print("Cluster Centers:\n",kmeans.cluster_centers_) 
    plt.figure(figsize=(7.5,5))
    plt.scatter(l[:,0],l[:,1], c=kmeans.labels_, cmap='rainbow')  
    plt.plot(kmeans.cluster_centers_[:,0] ,kmeans.cluster_centers_[:,1], 's', markerfacecolor='k', markersize=7,label="Center")
    plt.title('KMeans') 
    plt.legend(fancybox=False, framealpha=1, shadow=True, borderpad=1)
    plt.show()
    print(kmeans.labels_)  
    
k=int(input("k="))
l = [np.random.randint(0,100,2) for _ in range(100)]
l=np.reshape(l, (100,2))
k_means(k,l)
```
